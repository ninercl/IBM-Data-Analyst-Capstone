**Download and Load the Data**

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Dropping NaNs
df_valid = df.dropna(subset=['DatabaseWantToWorkWith'])

top5 = df_valid['DatabaseWantToWorkWith'].value_counts().head(5)
#print(top5)

#Pie Chart
top5.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Top 5 Databases Respondent Want to Work with')
plt.ylabel('')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Dropping NaNs
df_valid = df.dropna(subset=['DevType'])

top5_dev = df_valid['DevType'].value_counts().head(5)
#print(top5_dev)

#Pie Chart
top5_dev.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6))
plt.title('Top 5 Developer Roles')
plt.ylabel('') 
plt.show() 
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
#[col for col in df.columns if 'Op' in col or 'Sys' in col]
#Drop NaNs and create a Top N list
df_valid = df.dropna(subset=['OpSysProfessional use'])
top5_os = df_valid['OpSysProfessional use'].value_counts().head(5)

#PieCjart 
top5_os.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Operative System Used Professionally by Respondents') 
plt.ylabel('') 
plt.show() 
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
#df['LanguageHaveWorkedWith'].unique()
#Drop NaNs and Create a Top 5 List 
df_valid = df.dropna(subset=['LanguageHaveWorkedWith'])
top5_prog = df_valid['LanguageHaveWorkedWith'].value_counts().head(5)

#PieChart
top5_prog.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Top 5 Programming Languages Respondent Worked With')
plt.ylabel('') 
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
#Dropping NaNs 
df_valid = df.dropna(subset=['NEWCollabToolsHaveWorkedWith'])
top5_collab = df_valid['NEWCollabToolsHaveWorkedWith'].value_counts().head(5)

#Pie chart
top5_collab.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6))
plt.title('Top Collaboration Tools Used by Respondents') 
plt.ylabel('')
plt.show() 
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
#Dropping NaNs 
df_valid = df.dropna(subset=['LanguageAdmired'])
top5_admired = df_valid['LanguageAdmired'].value_counts().head(5)

#Pie Chart
top5_admired.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6))
plt.title('Top 5 Admired Language among Respondents')
plt.ylabel('') 
plt.show() 
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
#Drop Nans
df_valid = df.dropna(subset=['AIToolCurrently Using']) 

#Explode AI column 
df_exploded = df_valid['AIToolCurrently Using'].str.split(';').explode() 

top5_AI = df_exploded.value_counts().head(5) 

#Pie chart
top5_AI.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Top 5 AI Tool Currently Using')
plt.ylabel('') 
plt.show() 
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
#Drop Nans
df_valid = df.dropna(subset=['WebframeWantToWorkWith']) 

top5_fw = df_valid['WebframeWantToWorkWith'].value_counts().head(5)

#plot
top5_fw.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Top 5 Preferred Frameworks')
plt.ylabel('') 
plt.show() 
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
#Drop Nans
df_valid = df.dropna(subset=['EmbeddedWantToWorkWith']) 
df_exploded = df_valid['EmbeddedWantToWorkWith'].str.split(';').explode()

top5_emb = df_exploded.value_counts().head(5)

#Pie chart
top5_emb.plot(kind='pie', autopct='%1.1f%%', figsize=(8,6)) 
plt.title('Top 5 Embedded Respondents Want to Work With')
plt.ylabel('') 
plt.show() 
```
