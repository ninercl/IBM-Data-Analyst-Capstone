### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
#Drop Nans
df_valid = df.dropna(subset=['ConvertedCompYearly', 'Age'])
#Calculate medians
medians = df_valid.groupby('Age')['ConvertedCompYearly'].median()

#Plot 
medians.plot(kind='line', marker='o', figsize=(10,6)) 
plt.title('Median Yearly Compensation by Age')
plt.xlabel('Age')
plt.ylabel('Median Compensation') 
plt.xticks(rotation=45) 
plt.grid(True) 
plt.show() 
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
#Drop Nans
df_valid = df.dropna(subset=['ConvertedCompYearly', 'Age'])
#Filter Age 
df_filtered = df_valid[df_valid['Age'].isin(['25-34 years old', '35-44 years old'])]
#Calculate medians
medians = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

#Plot
medians.plot(kind='line', marker='o', figsize=(10,6)) 
plt.title('Median Compensation for Age (25-45)')
plt.xlabel('Age') 
plt.ylabel('Median Compensation')
plt.grid(True)
plt.show() 
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
# Drop NaNs
df_valid = df.dropna(subset=['JobSatPoints_6', 'YearsCodePro']).copy()

# Conver string values to numeric
def convert_years(x):
    if x == 'Less than 1 year':
        return 0
    elif x == 'More than 50 years':
        return 51
    else:
        return pd.to_numeric(x, errors='coerce')

df_valid['YearsCodePro_num'] = df_valid['YearsCodePro'].apply(convert_years)

# Calculate median
medians = df_valid.groupby('YearsCodePro_num')['JobSatPoints_6'].median()

#plot 
medians.plot(kind='line', marker='o', figsize=(10,6)) 
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Years of Professional Coding')
plt.ylabel('Median Job Satisfaction')
plt.grid(True) 
plt.xticks(rotation=45) 
plt.show() 
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
#Drop Nans
df_valid = df.dropna(subset=['ConvertedCompYearly', 'WorkExp'])
#Calculate medians
medians = df_valid.groupby('WorkExp')['ConvertedCompYearly'].median()

medians.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Median Yearly Compensation by Work Exp')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Compensation')
plt.grid(True)
plt.show() 
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
#Drop Nans
df_valid = df.dropna(subset=['JobSatPoints_6', 'WorkExp'])
#Calculate medians
medians = df_valid.groupby('WorkExp')['JobSatPoints_6'].median()
#plot
medians.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Median Job Satisfaction(6) by Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Job Satisfaction') 
plt.xticks(rotation=45) 
plt.grid(True)
plt.show() 
```
