### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
#Drop Nans
df_valid = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7', 'Age'])
#Calculate medians
medians = df_valid.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

#Graph
medians.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Median Job Satisfaction by Age Group') 
plt.ylabel('Median Score') 
plt.show() 
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
#Drop Nans
df_valid = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7', 'Employment'])
#Explode Employment
df_exploded = df_valid.assign(**{
    'Employment': df_valid['Employment'].str.split(';')
}).explode('Employment')

#Group by median
medians = df_exploded.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median()

#Graph 
medians.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Median Job Satisfcation by EmploymentStatus') 
plt.ylabel('Median Score') 
plt.show() 
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
#Drop Nans
df_valid = df.dropna(subset=['JobSatPoints_6', 'ConvertedCompYearly', 'Age'])

# Filter ages including 30-35
df_filtered = df_valid[df_valid['Age'].isin(['25-34 years old','35-44 years old'])]

#Group 
medians = df_filtered.groupby('Age')[['JobSatPoints_6', 'ConvertedCompYearly']].median()

#Graph 
medians_norm = medians.div(medians.max()) 
medians_norm.plot(kind='bar', stacked=True, figsize=(10,6)) 
plt.title('Median Compensation and Jobs Satisfaction (Age 25-34)') 
plt.ylabel('Median')
plt.show() 
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
#Drop Nans
df_valid = df.dropna(subset=['JobSat', 'Age'])

# Group by age and calculate medians
medians = df_valid.groupby('Age')[['JobSat', 'ConvertedCompYearly']].median()

#Median normalization 
medians_norm = medians.div(medians.max()) 

# Plot
medians_norm.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Median Compensation and Job Satisfaction Across Age Groups')
plt.ylabel('Median Values')
plt.xticks(rotation=45)
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
# Dropping NaNs
df_valid = df.dropna(subset=['DatabaseWantToWorkWith', 'Age']).copy()

# Calculate Top 5 most frequent DBs
top_dbs = df_valid['DatabaseWantToWorkWith'].value_counts().head(5).index

# Create a new column for 'other' DB
df_valid.loc[:, 'DatabaseGroup'] = df_valid['DatabaseWantToWorkWith'].where(
    df_valid['DatabaseWantToWorkWith'].isin(top_dbs), 'Other'
)

# Create table of proportions
table = pd.crosstab(df_valid['Age'], df_valid['DatabaseGroup'], normalize='index')

# Plot
table.plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab20')

plt.title('Preferred Databases by Age Group')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Database')
plt.show()


```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
# Drop NANs
df_valid = df.dropna(subset=['JobSat', 'Employment'])

# Explode 
df_exploded = df_valid.assign(Employment=df_valid['Employment'].str.split(';')).explode('Employment')

# table of frequency
table = pd.crosstab(df_exploded['Employment'], df_exploded['JobSat'], normalize='index')

# plot
table.plot(kind='bar', stacked=True, figsize=(12,6), cmap='tab20')
plt.title('Employment Type Across Job Satisfaction')
plt.ylabel('Proportion')
plt.xticks(rotation=45,ha='right', fontsize=9)
plt.show()

```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
# Drop NANs
df_valid = df.dropna(subset=['LanguageAdmired', 'Age'])

# Explode langauge admired
df_exploded = df_valid.assign(LanguageAdmired=df_valid['LanguageAdmired'].str.split(';')).explode('LanguageAdmired')

# Take top 10 most admired languages
top_langs = df_exploded['LanguageAdmired'].value_counts().head(10).index
df_filtered = df_exploded[df_exploded['LanguageAdmired'].isin(top_langs)]

# Frequency table
table = pd.crosstab(df_filtered['LanguageAdmired'], df_filtered['Age'], normalize='index')

# Plot
table.plot(kind='bar', stacked=True, figsize=(12,6), cmap='tab10')
plt.title('Technology Preference Across Age Groups')
plt.ylabel('Proportion')
plt.xticks(rotation=30, ha='right', fontsize=9)
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
# Drop NaNs
df_valid = df.dropna(subset=['PlatformAdmired', 'Employment'])

# Explode
df_exploded = (
    df_valid
    .assign(
        PlatformAdmired=df_valid['PlatformAdmired'].str.split(';'),
        Employment=df_valid['Employment'].str.split(';')
    )
    .explode('PlatformAdmired')
    .explode('Employment')
)

# Filter top 8 most admired platforms
top_platforms = df_exploded['PlatformAdmired'].value_counts().head(10).index
df_filtered = df_exploded[df_exploded['PlatformAdmired'].isin(top_platforms)]

# Table of frequency
table = pd.crosstab(df_filtered['PlatformAdmired'], df_filtered['Employment'], normalize='index')

# Plot
table.plot(kind='bar', stacked=True, figsize=(12,6), cmap='tab20')
plt.title('Platform Adoption by Employment Type')
plt.ylabel('Proportion')
plt.xticks(rotation=30, ha='right', fontsize=9)
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()

```
